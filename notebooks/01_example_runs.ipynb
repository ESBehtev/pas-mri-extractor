```python
from pas_mri_extractor.extractor import extract_mri_features
from pas_mri_extractor.scoring import normalize_mri_result
from pas_mri_extractor.extractor import extract_mri_features_with_raw
from pas_mri_extractor.models import load_llm
```

```python
loaded_model = load_llm()
```

```python
mri_text = """
Беременность 32 недели.

Плацента расположена преимущественно по передней стенке матки,
частично перекрывает внутренний зев.

В области послеоперационного рубца определяется выраженное
истончение миометрия.

Отмечаются множественные плацентарные лакуны и расширенные
ретроплацентарные сосуды.

Маточно-пузырное пространство местами не дифференцируется,
нельзя исключить вовлечение стенки мочевого пузыря.

Картина соответствует placenta increta.

Состояние после 2 кесаревых сечений.
"""
```

```python
debug = extract_mri_features_with_raw(
    mri_text,
    loaded_model=loaded_model,
)
print(debug["raw_output"])
```

```python
result = extract_mri_features(mri_text)
```

```python
final_result = normalize_mri_result(result)
```

```python
import json

print(
    json.dumps(
        final_result.model_dump(),
        ensure_ascii=False,
        indent=2,
    )
)
```
